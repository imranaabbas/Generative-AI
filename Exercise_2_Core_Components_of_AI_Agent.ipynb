# Exercise 2: Core Components of an AI Agent\n\nThis notebook helps you understand the **internal building blocks of an AI Agent**:\n- Brain (LLM)\n- Tools\n- Memory\n- Goals\n- Guardrails\n\n**Objective:** See how an agent is more than just a chatbot.

## 1️⃣ Agent Brain (LLM – Simulated)\nThe brain understands the goal and decides the next step.

```python
def agent_brain(goal):
    """Simulates the agent's brain, which understands the goal and decides the next step."""
    print(f"Agent Brain received goal: {goal}")
    return "Plan semester activities"

next_step = agent_brain("Plan my semester")
print("Next Step:", next_step)
```

## 2️⃣ Tools (Hands of the Agent)\nTools allow agents to **act**, not guess.

```python
def calendar_tool():
    """Simulates a tool that provides information, such as the start date of the semester."""
    return "Semester starts on July 1"

print("Tool Output:", calendar_tool())
```

## 3️⃣ Memory (Short-Term Memory)\nAgents remember context during execution.

```python
# Agent's memory to store context during execution
memory = {}
memory["semester"] = "2026-odd"
print("Agent Memory:", memory)
```

## 4️⃣ Goals & Instructions\nClear goals keep agents focused.

```python
# The main goal for the agent to achieve
goal = "Create a semester plan for a student"
print("Agent Goal:", goal)
```

## 5️⃣ Guardrails (Rules & Boundaries)\nGuardrails prevent unsafe or incorrect actions.

```python
# Guardrails define rules and boundaries for the agent's actions
guardrails = [
    "Do not modify official records",
    "Ask for approval before final submission"
]

print("Guardrails:")
for rule in guardrails:
    print("-", rule)
```
