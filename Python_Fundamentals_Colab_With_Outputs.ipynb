# 7. Flow of Execution

# This section explains how Python executes code step-by-step including function calls.
# Outcome: Trace execution paths and understand return values in multi-function programs.

# Execution starts at top, enters functions when called, returns on completion


```python
def add(a, b):
    result = a + b
    return result

x = add(5, 10)
print(x)

# More Examples:
def multiply(a, b):
    return a * b

def calculate(a, b):
    sum_val = add(a, b)
    product_val = multiply(a, b)
    return sum_val, product_val

result = calculate(3, 4)
print("Sum and Product:", result)

# ðŸ’ª Exercises:
# ðŸ§  Did You Know? Python executes top to bottom, and jumps into functions only when called. Understanding execution flow is critical to debugging.
# ðŸ”¹ 1. Trace flow with multiple function calls like add, multiply, divide
# 2. Create a billing system where total is calculated by calling multiple functions (tax, discount, net price)
# 3. Write a sequence of function calls to perform temperature conversion (Celsius to Fahrenheit)
```

# 8. Sequences - Lists, Tuples, Strings, Dictionaries

# Explore Python's collection types: lists (mutable), tuples (immutable), strings (text), dictionaries (key-value).
# Outcome: Access, modify, and iterate through sequences and mappings.

# Lists

```python
lst = [1, 2, 3]
print("List element at index 1:", lst[1])
lst.append(4)
print("Updated list:", lst)

# Tuples
tpl = (1, 2, 3)
print("Tuple element at index 2:", tpl[2])

# Strings
s = "Python"
print("First 3 characters:", s[:3])
print("Uppercase:", s.upper())

# Dictionary
d = {"name": "AI", "level": "Beginner"}
print("Name from dictionary:", d["name"])

# More Examples:
# List slicing and iteration
print("Sliced list (1:3):", lst[1:3])
for item in lst:
    print("List item:", item)

# String iteration
for char in s:
    print("Char:", char)

# Dictionary iteration
for key, value in d.items():
    print(f"{key} -> {value}")

# ðŸ’ª Exercises:
# ðŸ§  Did You Know? Strings, lists, tuples, and dictionaries are called sequences because you can loop through them or access them by index.
# ðŸ”¹ 1. Add 5 elements to a list and print them using a loop
# 2. Create a tuple of cities and print each city
# 3. Count vowels in a string using a loop
# 4. Update a dictionary with a new key-value pair and print it
```

# 9. Iteration (for, while)

# Learn how to automate repetitive tasks using loops and control their behavior using break and continue.
# Outcome: Write for and while loops to perform repeated tasks and navigate data structures.

# for loop

```python
for i in range(5):
    print(i)

# while loop
n = 0
while n < 3:
    print(n)
    n += 1

# More Examples:
# Using break
for i in range(10):
    if i == 5:
        break
    print("Break at 5:", i)

# Using continue
for i in range(5):
    if i == 2:
        continue
    print("Skip 2:", i)

# Nested loop
for i in range(1, 4):
    for j in range(1, 3):
        print(f"i={i}, j={j}")

# ðŸ’ª Exercises:
# ðŸ§  Did You Know? Python's `range()` function is memory-efficientâ€”it uses a generator instead of creating a full list in memory.
# ðŸ”¹ 1. Print all even numbers from 1 to 20 using a loop
# 2. Use a while loop to sum numbers until total exceeds 100
# 3. Create multiplication table of a number using a for loop
# 4. Use nested loops to print a pattern of stars (*)
```

# 11. Exception Handling

# Handle runtime errors gracefully using try-except blocks, and ensure cleanup with finally.
# Outcome: Make your programs robust and user-friendly by anticipating common runtime issues.

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
finally:
    print("Execution complete")

# More Examples:
try:
    num = int(input("Enter a number: "))
    print("You entered:", num)
except ValueError:
    print("Invalid input. Please enter a valid integer.")

# ðŸ’ª Exercises:
# ðŸ§  Did You Know? Python exceptions are objects that represent an error. You can even define your own by extending the Exception class.
# ðŸ”¹ 1. Handle input conversion error using try-except
# 2. Use exception handling to check file not found error
# 3. Create a calculator function that handles division by zero
# 4. Write code that uses finally to always close a file after reading
```

# 12. File Handling

# Learn how to read from and write to files using Pythonâ€™s open/read/write methods.
# Outcome: Store and retrieve data from text files and handle content programmatically.

# Writing to file

```python
with open("sample.txt", "w") as f:
    f.write("Hello from Python")

# Reading from file
with open("sample.txt", "r") as f:
    content = f.read()
    print(content)

# More Examples:
names = ["Alice", "Bob", "Charlie"]
with open("names.txt", "w") as f:
    for name in names:
        f.write(name + "
")

# Read and print each name
with open("names.txt", "r") as f:
    for line in f:
        print("Read name:", line.strip())

# ðŸ’ª Exercises:
# ðŸ§  Did You Know? File handling uses file pointers. When using `with open(...)`, Python automatically closes the file after exiting the block.
# ðŸ”¹ 1. Write a list of names to file and read it back
# 2. Count number of lines in a file
# 3. Read a file and count how many times a word appears
# 4. Write a program that logs user inputs into a log file
```

# 13. Data Structures - Stack and Queue

# Implement linear data structures using Python lists to simulate stacks (LIFO) and queues (FIFO).
# Outcome: Understand basic data structure operations and where they are used.

# Stack using List

```python
stack = []
stack.append(1)
stack.append(2)
print("Popped:", stack.pop())

# Queue using List
queue = []
queue.append(1)
queue.append(2)
print("Dequeued:", queue.pop(0))

# More Examples:
def push(stack, element):
    stack.append(element)
    print("Stack after push:", stack)

def pop(stack):
    if stack:
        print("Popped item:", stack.pop())
    else:
        print("Stack is empty")

def enqueue(queue, element):
    queue.append(element)
    print("Queue after enqueue:", queue)

def dequeue(queue):
    if queue:
        print("Dequeued item:", queue.pop(0))
    else:
        print("Queue is empty")

# ðŸ’ª Exercises:
# ðŸ§  Did You Know? Python lists can be used as stacks and queues, but for efficiency in large-scale applications, `deque` from the `collections` module is better.
# ðŸ”¹ 1. Implement push, pop for stack and enqueue, dequeue for queue
# 2. Simulate a ticket counter queue (enqueue names and dequeue as served)
# 3. Use stack logic to reverse a string
# 4. Create a menu-driven program to perform stack operations
```
